### 本金攤還的計算方式

```python
amount=100000
years=5
rate=2

amount,years,rate
```

```python
# 固定本金
period=years*12
monthly_pay=int(amount/period)  #round會無條件進位/int會無條件捨棄
monthly_pay
```

```python
#第一個月利息
monthly_rate=rate/100/12
monthly_rate
```

```python
interest=round(amount*monthly_rate)
interest
```

```python
#本利和
monthly_int=monthly_pay+interest
monthly_int
```

#### 每個月還款金額+利息

```python
amount=100000
Monthly=[]
total_int=0
for i in range(period):
    interest=round(amount*monthly_rate)
    monthly_int=monthly_pay+interest
    amount-=monthly_pay
    #last period:
    if i != (period-1):
        total_int+=interest
        Monthly.append([i+1,monthly_pay,interest,monthly_int,amount])
        print(f'期別:{i+1}, 本金:{monthly_pay}, 利息:{interest}, 本利和:{monthly_int}, 餘額:{amount},{total_int}')
    else:
        total_int+=interest
        Monthly.append([i+1,monthly_pay+amount,interest,monthly_pay+amount+interest,0])
        print(f'期別:{i+1}, 本金:{monthly_pay+amount}, 利息:{interest}, 本利和:{monthly_pay+amount+interest}, 餘額:{0},{total_int}')
print(Monthly)
```
